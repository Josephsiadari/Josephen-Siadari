```python
x = data.data
y = data.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=50)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi: {accuracy* 100:.2f}%")
```

```python
  from sklearn.datasets import load_digits
  from sklearn.tree import DecisionTreeClassifier, plot_tree
  import matplotlib.pyplot as plt

  # 1. Load dataset load_digits
  digits = load_digits()

  # 2. Pisahkan data dan target
  X = digits.data
  y = digits.target

  # 3. Latih model DecisionTreeClassifier
  model = DecisionTreeClassifier(random_state=50)
  model.fit(X, y)

  # 4. Plot pohon keputusan
  plt.figure(figsize=(40,20))  # Ukuran gambar
  plot_tree(model, filled=True, feature_names=[f'Pixel {i}' for i in range(X.shape[1])], class_names=[str(i) for i in digits.target_names])
  plt.show()
```
